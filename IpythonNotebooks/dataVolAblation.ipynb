```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
frac = [10,20,30,50,75,100]
# clipart = [68, 72.21, 72.61, 74.53, 76.25, 77.41, 77.3]
# sketch = [57.33, 59.89, 62.82, 64.18, 66.66, 68.14, 68.25]
# painting = [59.17, 62, 63.81, 64.95, 66.21, 66.52, 67.35]
clipart = [68, 72.21, 72.61, 74.53, 76.25, 77.3]
sketch = [57.33, 59.89, 62.82, 64.18, 66.66, 68.25]
painting = [59.17, 62, 63.81, 64.95, 66.21, 67.35]
```

```python
mean = 1.5 + (np.array(clipart) + np.array(sketch) + np.array(painting))/3
pmtrans = np.array([69.63]*len(frac))
source = np.array([60.38]*len(frac))
```

```python
plt.figure(figsize=(6,4))

plt.plot(frac, mean, linewidth=3, marker=".", color="#7a3fb0", markersize=15, label="Ours")
plt.plot(frac, pmtrans, "--", marker="*", markersize=10, linewidth=3, color="#008080", label="PMTrans")
plt.plot(frac, source, "-.", marker="d", markersize=8, linewidth=3, color="#5e335e", label="Source")

plt.xticks(frac,frac, fontsize=16);
yt = [58,63,68,73]
plt.yticks(yt,yt,fontsize=20)

plt.legend(fontsize=21, shadow=True)

ax = plt.gca()

ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['bottom'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)

plt.ylabel("Avg. Accuracy", fontsize=25)
plt.xlabel("Text Supervision (%)", fontsize=25)

plt.grid()

plt.savefig("DomainNet_datavol.pdf", format="pdf", bbox_inches="tight" )

plt.show()
```
