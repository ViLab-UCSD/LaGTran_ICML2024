## convert pseudo-labels to training_text

```python
import json
import pandas as pd
import numpy as np
import os
```

```python
officehome = json.load(open("../metadata/domainnet.json"))
```

```python
# mapping = {
#     "real" : "Real_World",
#     "art"  : "Art",
#     "clipart" : "Clipart",
#     "product" : "Product"
# }

all_d = ["real", "clipart", "sketch", "painting"]
mapping = {
    k:k for k in all_d
}

for domain in all_d:
    image_data = officehome[f"{domain}_train"]
    fid_to_impath = {int(im["id"]):os.path.join("/newfoundland/tarun/datasets/Adaptation/OfficeHome/Dataset10072016/", im["filename"]) for im in image_data["images"]}
    for src in ["real", "art", "clipart", "product"]:
        if src == domain:
            continue
        pseudo = pd.read_csv(f"../hard_labels/officehome_{src}_{domain}_blipPL.txt", sep="\s", engine="python", header=None, names=["a","b"])
        pseudo["filepath"] = pseudo.a.map(fid_to_impath)
        
        with open(f"/home/tarun/DomainAdaptationModels/PMTrans_original/dataset/office_home/{mapping[src]}_{mapping[domain]}_pseudo.txt", "w") as fh:
            write_str = ""
            for path,label in zip(pseudo.filepath, pseudo.b):
                write_str += f"{path} {label}\n"
            fh.write(write_str)
```
